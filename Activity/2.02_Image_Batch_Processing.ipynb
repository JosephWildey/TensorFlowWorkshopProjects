# Activity 2.02: Loading Image Data for Batch Processing

Load images for batch processing and adjust the input data with random pertubations such as rotations, flipping the image horizontally, and adding shear to the image. This will create additional training data.

## Import tensorflow

```python
import tensorflow as tf
```

## Import ImageDataGenerator

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

training_datagen = ImageDataGenerator(rescale=1./255,
                               shear_range=0.2,
                               rotation_range=180,
                               zoom_range=0.2,
                               horizontal_flip=True)
```

## Import image data

```python
training_set = training_datagen.flow_from_directory("C:\\Users\\JoeWi\\TensorFlowWorkshopProjects\\Datasets\\image_data",
                                                target_size = (64,64),
                                                batch_size = 25,
                                                class_mode = 'binary')
```

## Display Images in the Batch

```python
import matplotlib.pyplot as plt 

def show_batch(image_batch, label_batch):
    
    lookup = {v: k for k, v in training_set.class_indices.items()}
    
    label_batch = [lookup[label] for label in label_batch]
    
    plt.figure(figsize=(10,10))
    
    for n in range(25):
        
        ax = plt.subplot(5, 5, n+1)
        
        plt.imshow(image_batch[n])
        
        plt.title(label_batch[n].title())
        
        plt.axis('off')
```

## Feed images into the function above

```python
image_batch, label_batch = next(training_set)

show_batch(image_batch, label_batch)
```
