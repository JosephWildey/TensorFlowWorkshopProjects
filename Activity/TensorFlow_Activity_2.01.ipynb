# Activity 2.01: Loading Tabular Data and Rescaling Numerical Fields with MinMax Scaler

The information used in this activity is teperature data from Seoul, Korea.

## Import Pandas

```python
import pandas as pd
```

## Import tabular data into Pandas Dataframe

```python
df = pd.read_csv("C:\\Users\\JoeWi\\The-TensorFlow-Workshop\\Chapter02\\Datasets\\Bias_correction_ucl.csv")

df
```

## Drop the "Date" column

```python
df.drop("Date", inplace=True, axis=1)
```

## Plot histogram

```python
ax = df['Present_Tmax'].hist(color='gray')
ax.set_xlabel("Temperature")
ax.set_ylabel("Frequency")
```

## Import and apply minmax scaler

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df2 = scaler.fit_transform(df)

df2 = pd.DataFrame(df2, columns=df.columns)
```

## Plot new Histogram

Values will be between 0 and 1 due to the use of the MinMax Scaler

```python
ax = df2["Present_Tmax"].hist(color="gray")
ax.set_xlabel("Temperature")
ax.set_ylabel("Frequency")
```
