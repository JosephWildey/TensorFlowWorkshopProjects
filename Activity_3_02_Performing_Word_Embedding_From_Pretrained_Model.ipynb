Import tensorflow and tensorflow hub

```python
import tensorflow as tf
import tensorflow_hub as hub
```

Assign the location of a pretrained model as the model_handle

```python
model_handle = "https://tfhub.dev/google/universal-sentence-encoder/4"
```

Create a KerasLayer for the model, with the pretrained model, an empty list for the input shape, and its type as a string

```python
hub_layer = hub.KerasLayer(model_handle, input_shape=[], dtype=tf.string)
```

Create a list for embedding with the model

```python
text = ['The TensorFlow Workshop']
```

Apply the text to the model

```python
hub_layer(text)
```
