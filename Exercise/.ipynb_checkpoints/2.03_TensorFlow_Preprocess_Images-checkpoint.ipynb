## Import tensorflow

## Import the DataGenerator class

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

## Instatiate the DataGenerator Class Creator

By passing the argument 1./255 to rescale, it will make the image values between 0 and 1.

```python
train_datagen = ImageDataGenerator(rescale = 1./255)
```

## Use flow from directory to direct the data generator to the data

```python
training_set = train_datagen.flow_from_directory('C:\\Users\\JoeWi\\TensorFlowWorkshopProjects\\Datasets\\image_data',
                                               target_size = (64,64),
                                               batch_size = 25,
                                               class_mode = 'binary')
```

## Display the images in the batch

```python
import matplotlib.pyplot as plt

def show_batch(image_batch, label_batch):
    
    lookup = {v: k for k, v in training_set.class_indices.items()}
    
    label_batch = [lookup[label] for label in label_batch]
    
    plt.figure(figsize=(10,10))
    
    for n in range(25):
        
        ax = plt.subplot(5,5,n+1)
        plt.imshow(image_batch[n])
        plt.title(label_batch[n].title())
        plt.axis('off')
```

```python
image_batch, label_batch = next(training_set)

show_batch(image_batch, label_batch)
```
